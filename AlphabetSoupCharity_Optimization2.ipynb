### Deliverable 1: Preprocessing the Data for a Neural Network

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN', 'NAME', 'STATUS'])
```

```python
# Determine which values to replace if counts are less than ...?
AT_counts = application_df.APPLICATION_TYPE.value_counts()
replace_application = list(AT_counts[AT_counts < 500].index)

# Replace in dataframe
for app in replace_application:
    application_df.APPLICATION_TYPE = application_df.APPLICATION_TYPE.replace(app,"Other")
    
# Check to make sure binning was successful
application_df.APPLICATION_TYPE.value_counts()
```

```python
# Determine which values to replace if counts are less than ..?
class_counts = application_df.CLASSIFICATION.value_counts()
replace_class = list(class_counts[class_counts <= 100].index)

# Replace in dataframe
for cls in replace_class:
    application_df.CLASSIFICATION = application_df.CLASSIFICATION.replace(cls,"Other")
    
# Check to make sure binning was successful
application_df.CLASSIFICATION.value_counts()
```

```python
# Generate our categorical variable lists
cvars = application_df.dtypes[application_df.dtypes == 'object'].index.tolist()
cvars
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(application_df[cvars]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(cvars)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
application_df = application_df.merge(encode_df, left_index = True, right_index = True).drop(columns = cvars)
application_df
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL']
X = application_df.drop(columns = ['IS_SUCCESSFUL']).values

# Split the preprocessed data into a training and testing dataset
ttot = .75
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = ttot, random_state = 78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Deliverable 2: Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 8
hidden_nodes_layer2 = 5
hidden_nodes_layer3 = 3

model = tf.keras.models.Sequential()

# First hidden layer
model.add(tf.keras.layers.Dense(units = hidden_nodes_layer1, 
        input_dim = number_input_features, activation = 'relu'))

# Second hidden layer
model.add(tf.keras.layers.Dense(units = hidden_nodes_layer2, 
        activation = 'relu'))

#Third hidden layer
model.add(tf.keras.layers.Dense(units = hidden_nodes_layer3, 
        activation = 'relu'))

# Output layer
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Check the structure of the model
model.summary()
```

```python
# Compile the model
model.compile(loss = 'binary_crossentropy', 
                optimizer = 'adam', metrics = ['accuracy'])
```

```python
#Add checkpoints
import os
from tensorflow.keras.callbacks import ModelCheckpoint

os.makedirs('checkpoints2/', exist_ok = True)
checkpoint_path = 'checkpoints2/weights.{epoch:02d}.hdf5'

#Create callback
cp_callback = ModelCheckpoint(
    filepath = checkpoint_path,
    verbose = 1,
    save_weights_only = True,
    save_freq = 804*5)
```

```python
# Train the model
fit_model = model.fit(X_train, y_train, epochs = 100, 
                      callbacks = [cp_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
#Export model
model.save('AlphabetSoupCharity_Optimization2.h5')
```
